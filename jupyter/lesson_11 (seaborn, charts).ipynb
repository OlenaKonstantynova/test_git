```python
plt.subplots(figsize = (8,6))
sns.scatterplot(x = 'total_bill', y = 'tip', data = tips, hue = 'smoker');
plt.grid()
```

```python
plt.subplots(figsize = (8,6))
sns.scatterplot(x = 'total_bill', y = 'tip', data = tips, style = 'time', hue = 'smoker');
```

```python
sns.relplot(x = 'total_bill', y = 'tip', col = 'time' ,hue = 'smoker', data = tips);
```

```python
sns.relplot(x = 'total_bill', y = 'tip', col = 'time', hue = 'size', data = tips);
```

```python
sns.relplot(x = 'total_bill', y = 'tip', col = 'time', hue = 'size', data = tips, palette  = sns.cubehelix_palette(rot =-.4),);
```

```python
tips.groupby('size').agg(sum)
```

```python
tips['size'].value_counts()
```

```python
plt.figure(figsize = (10,6))

sns.set_style('whitegrid')  # stile of chart (in seaborn)

sns.swarmplot(x= 'day', y ='total_bill', data = tips);
plt.xlabel('Day of week', )
plt.ylabel('Total bill')

plt.show()
```

```python
plt.figure(figsize = (10,6))

sns.set_style('whitegrid')

sns.pointplot(x= 'day', y ='total_bill', hue = 'sex',  data = tips);

plt.xlabel('Day' )
plt.ylabel('Total bill')

plt.show()
```
