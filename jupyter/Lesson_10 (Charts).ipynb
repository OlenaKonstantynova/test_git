```python
x1 = [1,2,3,4]
x2 = [5,6,7,8]

x3 = [1,2,3,7]
x4 = [5,6,4,8]

x5 = [1,2,3,4]
x6 = [i ** 3 for i in x5]


plt.plot(x1,x2)
plt.plot(x3,x4)
plt.plot(x5,x6)

plt.title('Second graph', {'fontname': 'Bradley Hand', 'fontsize': 15}) # дать название графику
plt.xlabel('x - axis', {'fontname': 'Skia', 'fontsize': 12}) # дать название оси x 
plt.ylabel('y - axis', {'fontname': 'Arial Black', 'fontsize': 12}) 
```

```python
x1 = [1,2,3,4]
x2 = [5,6,7,8]

x3 = [1,2,3,7]
x4 = [5,6,12,16]

x5 = [1,2,3,4]
x6 = [i ** 3 for i in x5]


plt.plot(x1,x2, label = 'linear', color = '#87d4fc')
plt.plot(x3,x4, label = 'curve', color = 'green')
plt.plot(x5,x6, label = 'pow in 3', color = 'black', linestyle = '--')

plt.title('Third graph', {'fontname': 'Bradley Hand'}) # дать название графику
plt.xlabel('x - axis', {'fontname': 'Skia'}) # дать название оси x 
plt.ylabel('y - axis', {'fontname': 'Arial Black'}) 
plt.legend()
```

```python
x1 = [1,2,3,4]
x2 = [5,6,7,8]

x3 = [1,2,3,7]
x4 = [5,6,12,16]

x5 = [1,2,3,4]
x6 = [i ** 3 for i in x5]


plt.plot(x1,x2, label = 'linear', color = '#87d4fc', linewidth = 3, linestyle = 'solid', marker = 'o')
plt.plot(x3,x4, label = 'curve', color = 'green', marker = '*', markersize = 10)
plt.plot(x5,x6, label = 'pow in 3', color = 'black', linestyle = '--')

plt.title('Fourth graph', {'fontname': 'Bradley Hand'}) # дать название графику
plt.xlabel('x - axis', {'fontname': 'Skia'}) # дать название оси x 
plt.ylabel('y - axis', {'fontname': 'Arial Black'}) 
plt.grid()
plt.legend()
```

```python
day = df_1.groupby('year').agg(total = ('births','sum'))
plt.plot(day, color = 'purple', linestyle = 'solid', linewidth = 4, marker = 'o', markersize = 7)

plt.title('Births by year', {'fontname': 'Skia', 'fontsize': 15}) 
plt.xlabel('x - axis', {'fontname': 'Skia', 'fontsize': 11})  
plt.ylabel('y - axis', {'fontname': 'Skia', 'fontsize': 11}) 

plt.grid()
plt.show()
```

```python
plt.plot(day,color = '#0B0A02', linewidth = 1, linestyle = '--', marker = '.', 
         markersize = 15, markerfacecolor = 'red')

plt.grid()
plt.title('Births by years', {'fontname': 'Tahoma', 'fontsize': 18}) 
plt.xlabel('x - year of birth',{'fontname': 'Skia','fontsize': 15}) 
plt.ylabel('y - births',{'fontname': 'Skia','fontsize': 15}) 

plt.grid(color = 'blue')
```
